# Ensembles

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import VotingClassifier
from sklearn.model_selection import train_test_split

from sklearn.datasets import make_moons, load_breast_cancer

X, y = make_moons(noise=.2, random_state=18) # carefully picked random state for illustration
y_label = np.take(labels, (y > 0.5))

X_train, X_test, y_train, y_test, y_label, _ = train_test_split(X, y, y_label,
                                                                stratify=y, random_state=0)

voting = VotingClassifier([('logreg', LogisticRegression(C=100)),
                           ('tree', DecisionTreeClassifier(max_depth=3, random_state=0))],
                          voting='soft')
voting.fit(X_train, y_train)

lr, tree = voting.estimators_

print(("{:.2f} " * 3).format(voting.score(X_test, y_test),
                             lr.score(X_test, y_test),
                             tree.score(X_test, y_test)))
```

```python
plot_surface(lr, X_train, y_label, show=False)
plt.title("Logistic Regression");
plt.xlabel("Feature 1")
plt.ylabel("Feature 2");
plt.savefig("../../images/voting_lr.png");
```

# Gradient boosting illustrated

```python
def f(x):
    return x + np.sin(x) + np.sin(3 * x)

line = np.linspace(0, 10, 200)
x = np.linspace(0, 10, 200)
rng = np.random.RandomState(0)
rng.shuffle(x)
X = np.sort(x[:40])
y = f(X)

X = X.reshape(-1, 1)
```

```python
plt.plot(line, f(line), '--r', label='truth')
plt.plot(X, y, 'ob', label='observations')
plt.xlabel('x')
plt.ylabel('$f(x)=x+\sin(x)+\sin(3x)$')
plt.legend(loc='best');
plt.savefig("../../images/gradient_boost_true.png")
```

```python
from sklearn.ensemble import GradientBoostingRegressor

rgr = GradientBoostingRegressor(n_estimators=3, learning_rate=1)
rgr.fit(X, y)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 8))

ax.plot(line, f(line), '--r', label='truth (unknown)')
ax.plot(X, y, 'ob', label='data')
ax.plot(line, rgr.predict(line.reshape(-1, 1)), '-g', label='full model')
ax.set_ylabel('f(x)')
ax.set_xlabel("X")
ax.legend(loc='best');
fig.savefig("../../images/gbr_data.png")
```

```python
prev_prediction = np.ones_like(y) * np.mean(y)

fig, ax = plt.subplots(1, 1, figsize=(8, 8))
ax.plot([0, 10], prev_prediction[0:2])
ax.plot(X, y, 'ob', label='data')
ax.set_xlabel("X")

# "boosts" or gradient steps
for n, est in enumerate(rgr.estimators_):
    fig, ax = plt.subplots(1, 1, figsize=(8, 8))

    ax.plot(line, est[0].predict(line.reshape(-1, 1)))
    ax.plot(X, y - prev_prediction, 'xb')
    prev_prediction += est[0].predict(X)
    ax.set_xlabel("X")
    ax.set_ylabel("residual")
    fig.savefig("../../images/gbr_stage_residuals_%i.png" % (n+1))
```

```python
prev_prediction = np.ones_like(y) * np.mean(y)

fig, ax = plt.subplots(1, 1, figsize=(8, 8))
ax.plot([0, 10], prev_prediction[0:2], label='stage 0')
ax.plot(X, y, 'ob', label='data')
ax.set_xlabel("X")
ax.set_ylabel("f(x)")
ax.legend(loc='best')
fig.savefig("../../images/gbr_stage_0.png")

# "boosts" or gradient steps
for n, prediction in enumerate(rgr.staged_predict(line.reshape(-1, 1))):
    fig, ax = plt.subplots(1, 1, figsize=(8, 8))

    ax.plot(line, prediction, label='stage %i' % (n+1))
    ax.plot(X, y, 'ob', label='data')

    ax.set_xlabel("X")
    ax.set_ylabel("f(x)")
    ax.legend(loc='best')
    fig.savefig("../../images/gbr_stage_%i.png" % (n+1))
```

# Stacking

```python
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier


labels = ["b", "r"]
X, y = make_blobs(n_samples=400, centers=23, random_state=42)
y_label = np.take(labels, (y < 10))
y = np.take([0, 1], (y < 10))
X_train, X_test, y_train, y_test, y_label,_ = train_test_split(X, y, y_label,
                                                               random_state=0)

voting = VotingClassifier([('logreg', LogisticRegression(C=100)),
                           ('tree', DecisionTreeClassifier(max_depth=5, random_state=0)),
                           # for an even more impressive overfit set this to 1
                           # and look at the LogisticRegression weights later on!
                           ('knn', KNeighborsClassifier(n_neighbors=3))
                          ],
                         voting='soft', flatten_transform=True)
voting.fit(X_train, y_train)
lr, tree, knn = voting.estimators_
print(("{:.2f} " * 4).format(voting.score(X_test, y_test),
                             lr.score(X_test, y_test),
                             tree.score(X_test, y_test),
                             knn.score(X_test, y_test)))
```

```python
x_lin = np.linspace(X_train[:, 0].min() - .5, X_train[:, 0].max() + .5, 100)
y_lin = np.linspace(X_train[:, 1].min() - .5, X_train[:, 1].max() + .5, 100)
x_grid, y_grid = np.meshgrid(x_lin, y_lin)
X_grid = np.c_[x_grid.ravel(), y_grid.ravel()]
# transform produces individual probabilities
_, lr_probs, _, tree_probs, _, knn_probs = voting.transform(X_grid).T
```

```python
titles = ['Logistic Regression', 'Decision Tree', 'KNN', 'Average']
for prob, title in zip([lr_probs, tree_probs, knn_probs, lr_probs + tree_probs + knn_probs],
                       titles):
    fig, ax = plt.subplots(1, 1, figsize=(8, 8))
    ax.set_title(title)
    ax.contourf(x_grid, y_grid, prob.reshape(x_grid.shape),
                alpha=0.8, cmap=plt.cm.RdBu_r)
    ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap='bwr')
    plt.show()
    fig.savefig("../../images/voting_%s.png" % title)
```

```python
from sklearn.preprocessing import FunctionTransformer
from sklearn.pipeline import make_pipeline


# keep only the probabilities for the positive class
stacking = make_pipeline(voting,
                         FunctionTransformer(lambda X: X[:, 1::2]),
                         LogisticRegression())
stacking.fit(X_train, y_train)
stacking.score(X_train, y_train)
```

```python
stacking.score(X_test, y_test)
```

```python
stacking.named_steps.logisticregression.coef_
```

This will select the classifier that memorises the training data the best. Why? Because the scores from that model will be the best for fitting the logistic regression.

The way out is to use a set of data that was not used during the fitting of the base models.

```python
from sklearn.model_selection import cross_val_predict

first_stage = make_pipeline(voting,
                            FunctionTransformer(lambda X: X[:, 1::2]))
transform_cv = cross_val_predict(first_stage, X_train, y_train, cv=10, method="transform")
```

```python
second_stage = LogisticRegression().fit(transform_cv, y_train)
print(second_stage.coef_)
```

```python
print("training score: %.2f" % second_stage.score(transform_cv, y_train))
print("testing score: %.2f" % second_stage.score(first_stage.transform(X_test), y_test))
# difference is much smaller, less overfitting
```
